```python
import pandas as pd
import numpy as np

# Different date ranges for each symbol
dates_aapl = pd.date_range(start='2024-01-01', end='2024-01-31', freq='D')
dates_googl = pd.date_range(start='2024-01-15', end='2024-02-15', freq='D')
dates_msft = pd.date_range(start='2024-02-01', end='2024-02-29', freq='D')
dates_amzn = pd.date_range(start='2024-01-10', end='2024-02-10', freq='D')
dates_meta = pd.date_range(start='2024-02-05', end='2024-03-05', freq='D')
dates_nvda = pd.date_range(start='2024-01-20', end='2024-02-20', freq='D')
dates = {
    'AAPL': dates_aapl,
    'GOOGL': dates_googl,
    'MSFT': dates_msft,
    'AMZN': dates_amzn,
    'META': dates_meta,
    'NVDA': dates_nvda
}

# Create sample symbols
symbols = ['AAPL', 'GOOGL', 'MSFT', 'AMZN', 'META', 'NVDA']

# Create sample data
data = []
for symbol in symbols:
    for date in dates[symbol]:
        # Generate random price and volume data
        if symbol == 'AAPL':
            price = np.random.uniform(150, 200)
        elif symbol == 'GOOGL':
            price = np.random.uniform(100, 150)
        elif symbol == 'MSFT':
            price = np.random.uniform(50, 100)
        elif symbol == 'AMZN':
            price = np.random.uniform(120, 170)
        elif symbol == 'META':
            price = np.random.uniform(200, 250)
        elif symbol == 'NVDA':
            price = np.random.uniform(400, 450)
        else:
            price = np.random.uniform(50, 100)
        volume = np.random.uniform(1000000, 5000000)
        data.append({
            'datetime': date,
            'symbol': symbol,
            'price': price,
            'volume': volume
        })

# Create DataFrame
df = pd.DataFrame(data)

# Display first few rows
print("Sample of the dataframe:")
print(df.head())


```

```python
# Create plots
plt = plot_column(df, 'price')
plt.show()

plt = plot_column(df, 'volume')
plt.show()
```
